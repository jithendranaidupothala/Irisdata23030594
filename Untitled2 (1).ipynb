```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# We have to load the Iris dataset
iris_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                        names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Display the first few rows of the dataset
print(iris_data.head())
```

```python
# Function to plot histogram of a specific feature
def plot_histogram(data, feature):


    plt.figure(figsize=(8, 6))
    sns.histplot(data[feature], bins=30, kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()


# Plot histogram of sepal length
plot_histogram(iris_data, 'sepal_length')
```

```python
# Function to create a bar graph showing average values of each numerical feature for each species
def plot_bar(data):
    plt.figure(figsize=(10, 6))
    iris_grouped = data.groupby('species').mean()
    iris_grouped.plot(kind='bar', colormap='viridis')
    plt.title('Average Values of Numerical Features for Each Species')
    plt.xlabel('Species')
    plt.ylabel('Average Value')
    plt.xticks(rotation=45)
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.show()
plot_bar(iris_data)
```

```python
def plot_line(data, feature):
    plt.figure(figsize=(8, 6))
    plt.plot(data[feature], marker='o', color='blue', linestyle='-')
    plt.title(f'Line Graph of {feature}')
    plt.xlabel('Index')
    plt.ylabel(feature)
    plt.grid(True)
    plt.show()
plot_line(iris_data, 'sepal_length')
```

```python
# Function to plot scatter plot of two features
def plot_scatter(data, feature1, feature2):


    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=feature1, y=feature2, data=data, hue='species')
    plt.title(f'Scatter Plot of {feature1} vs {feature2}')
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.show()

# Plot scatter plot of sepal length vs sepal width
plot_scatter(iris_data, 'sepal_length', 'sepal_width')
```

```python
# Basic statistics of the dataset
print(iris_data.describe())

# Correlation matrix of numerical variables
print(iris_data.corr())
```
